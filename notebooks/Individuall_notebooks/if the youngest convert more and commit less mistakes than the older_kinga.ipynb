# Mistake rate per age group
A mistake = pushback to the previous step in the process.

### Step 1: Calculate mistakes per visit

```python
# Sort by visit and datetime
df = df.sort_values(['visit_id', 'date_time'])

df
```

```python
# Map process steps to numerical order
step_order = {'start': 0, 'step_1': 1, 'step_2': 2, 'step_3': 3, 'confirm': 4}
df['step_num'] = df['process_step'].map(step_order)

```

```python
def count_mistakes(steps):
    """Count how many times the process moves backward in a visit."""
    mistakes = 0
    for i in range(1, len(steps)):
        if steps[i] < steps[i-1]:
            mistakes += 1
    return mistakes
```

```python
# Count mistakes per visit
mistakes_per_visit = df.groupby('visit_id')['step_num'].apply(lambda x: count_mistakes(list(x)))
df_mistakes = mistakes_per_visit.reset_index().rename(columns={'step_num': 'mistakes'})

df_mistakes
```

```python
# Step 2: Merge with client info

# Keep one row per visit with age_band info
df_visits = df[['visit_id', 'client_id', 'age_band']].drop_duplicates()

df_visits
```

```python
# Merge mistakes
df_visits = df_visits.merge(df_mistakes, on='visit_id', how='left')

df_visits
```

```python
# Step 3: Aggregate by age_band

age_analysis = df_visits.groupby('age_band').agg(
    total_visits=('visit_id', 'count'),
    total_mistakes=('mistakes', 'sum'),
    avg_mistakes=('mistakes', 'mean')
).reset_index()
```

```python
# Step 4: Display results

print(age_analysis)
```

```python
# Statistical check: mistake vs age_band
```

```python
from scipy.stats import f_oneway
```

```python
# Prepare mistake data per age_band
groups = [group['mistakes'].values for name, group in df_visits.groupby('age_band')]
```

```python
# One-way ANOVA
f_stat, p = f_oneway(*groups)
print(f"ANOVA F-statistic: {f_stat:.3f}, p-value: {p:.5f}")

if p < 0.05:
    print("The average number of mistakes significantly differs by age band.")
else:
    print("No significant difference in mistakes by age band.")
```

# Conversion rate & Mistake rate per age group

```python
# Merge conversion_rate & mistake_rate with visit-level info

df_visits_both_rates = df[['visit_id', 'client_id', 'age_band', 'converted']].drop_duplicates()
df_visits_both_rates = df_visits_both_rates.merge(df_mistakes, on='visit_id', how='left')

df_visits_both_rates
```

```python
# Aggregate by age_band

age_analysis_both = df_visits_both_rates.groupby('age_band').agg(
    total_visits_both=('visit_id', 'count'),
    total_converted_both=('converted', 'sum'),
    conversion_rate_both=('converted', 'mean'),
    total_mistakes_both=('mistakes', 'sum'),
    avg_mistakes_both=('mistakes', 'mean')
).reset_index()
```

```python
# Display results

print("Conversion rate and average mistakes per age band:")
print(age_analysis_both)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 'age_band', 'conversion_rate_both', 'avg_mistakes_both'

fig, ax1 = plt.subplots(figsize=(8,5))

# Bar plot for Conversion Rate
sns.barplot(
    x='age_band', y='conversion_rate_both', data=age_analysis_both,
    palette='Blues', ax=ax1
)
ax1.set_ylabel('Conversion Rate', color='blue')
ax1.set_xlabel('Age Band')
ax1.set_title('Conversion Rate and Average Mistakes per Age Band')

# Second y-axis for Average Mistakes
ax2 = ax1.twinx()
sns.lineplot(
    x='age_band', y='avg_mistakes_both', data=age_analysis_both,
    color='red', marker='o', linewidth=2, ax=ax2
)
ax2.set_ylabel('Average Mistakes', color='red')

# Optional: Show numeric values on bars/line
for idx, row in age_analysis_both.iterrows():
    ax1.text(idx, row['conversion_rate_both'] + 0.01, f"{row['conversion_rate_both']:.1%}", 
             ha='center', color='blue')
    ax2.text(idx, row['avg_mistakes_both'] + 0.02, f"{row['avg_mistakes_both']:.2f}", 
             ha='center', color='red')

plt.tight_layout()
plt.show()
```
